```python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.stats import norm
```

```python
ticker = 'SPY'
data = yf.download(ticker, start='2022-01-01', end='2025-10-15', progress=False)['Close']

daily_returns = data.pct_change().dropna()
sigma = daily_returns.std() * np.sqrt(252)

S0 = data.iloc[-1][0]
r = 0.0411
T = 100 / 365
K = 680
```

```python
d1 = (np.log(S0 / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)

call_black_scholes = S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
put_black_scholes = K * np.exp(-r * T) * norm.cdf(-d2) - S0 * norm.cdf(-d1)

print(f"Black–Scholes Call: ", call_black_scholes)
print(f"Black–Scholes Put: ", put_black_scholes)
```
