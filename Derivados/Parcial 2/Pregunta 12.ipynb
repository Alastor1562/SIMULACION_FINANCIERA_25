```python
import numpy as np

np.random.seed(666)
```

```python
# Parámetros de la opción
S0 = 120
K  = 100
r  = 0.07
sigma = 0.10
T = 1
n_sim = 100000

# Simulación con GBM
Z = np.random.normal(0,1,n_sim)
ST = S0 * np.exp((r - 0.5*sigma**2)*T + sigma*np.sqrt(T)*Z)

# Payoff call al vencimiento
payoff = np.maximum(ST - K, 0)

# Prima descontada a valor presente
price = np.exp(-r*T) * payoff.mean()
price
```
