```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del straddle
K = 50 
call = 5
put = 5
cost_total = call + put

ST = np.linspace(25, 75, 400)  # rango de precios para evaluar

# Payoff neto del straddle largo
payoff_call = np.maximum(ST - K, 0)
payoff_put  = np.maximum(K - ST, 0)
PL = payoff_call + payoff_put - cost_total

# Puntos de equkilibrio
breakeven_low  = K - cost_total
breakeven_high = K + cost_total

# Grfica
plt.figure()
plt.plot(ST, PL, label="Payoff neto del straddle largo", color="blue")
plt.title("Payoff neto del straddle largo")
plt.xlabel("Precio del subyacente al vencimiento $S_T$")
plt.ylabel("Ganancia/Prdida neta")

# Ejes
plt.axhline(0, linewidth=1, linestyle="--")
plt.axvline(K, linewidth=1, linestyle="--")

# Puntos de equilibrio
plt.axvline(breakeven_low, color="red", linestyle="--", label="Punto de equilibrio bajo")
plt.axvline(breakeven_high, color="green", linestyle="--", label="Punto de equilibrio alto")

plt.legend()
plt.grid()
plt.show()
```
